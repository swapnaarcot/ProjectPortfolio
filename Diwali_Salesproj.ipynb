```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali_Sales_Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax=sns.countplot(x='Gender',data=df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
plt.figure(figsize=(15,3))
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_amt=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False).head(10)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=sales_amt,x='State',y='Amount')
```

```python
ax=sns.countplot(data=df,x='Marital_Status')
for i in ax.containers:
    ax.bar_label(i)
    
```

```python
sales_mgen=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False).head(10)
sns.barplot(data=sales_mgen,x='Marital_Status',y='Amount',hue='Gender')
```

```python
plt.figure(figsize=(20,5))
ax=sns.countplot(data=df,x='Occupation')
for i in ax.containers:
    ax.bar_label(i)
```

```python
plt.figure(figsize=(20,5))
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_occ,x='Occupation',y='Amount')
```

```python
plt.figure(figsize=(25,5))
ax=sns.countplot(data=df,x='Product_Category')
for i in ax.containers:
    ax.bar_label(i)
```

```python
plt.figure(figsize=(25,5))
sales_cat=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=sales_cat,x='Product_Category',y='Amount')
```

```python
sales_prod=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
plt.figure(figsize=(15,5))
sns.barplot(data=sales_prod,x='Product_ID',y='Orders')
```
