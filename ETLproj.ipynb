```python
#Handle null values

df=pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
df.columns
```

```python
# Change the column names to std -lower and _
df.columns=df.columns.str.lower()
```

```python
df.columns=df.columns.str.replace(' ','_')
```

```python
df.head(10)
```

```python
# Add a new column discount column
df['discount']=df['list_price']*df['discount_percent']*0.01
```

```python
df.head(5)
```

```python
df['sale_price']=df['list_price']-df['discount']
```

```python
df.head(4)
```

```python
df['profit']=df['sale_price']-df['cost_price']
df.head(5)
```

```python
#Change the data type of Order date from object to date

df.dtypes
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
```

```python
df.dtypes
```

```python
# Drop irrelevant columns
df.drop(columns=['list_price','cost_price','discount'],inplace=True)    
df
```
