**Finding outliers**

```python
# Outliers are less than Q1-1.5IQR and greater than Q3+1.5IQR

Q1=np.percentile(bank_df['Income'],25)
Q3=np.percentile(bank_df['Income'],75)
IQR=Q3-Q1
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR
con1=bank_df['Income']<lb
con2=bank_df['Income']>ub

outliers=bank_df[con1|con2]
count=len(outliers)
count

```

```python
#Non-outliers data
Q1=np.percentile(bank_df['Income'],25)
Q3=np.percentile(bank_df['Income'],75)
IQR=Q3-Q1
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR
con1=bank_df['Income']>lb
con2=bank_df['Income']<ub
non_outliers=bank_df[con1&con2]
count=len(non_outliers)
count
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(bank_df['Income'])
plt.subplot(1,2,1).hist(non_outliers['Income'])
plt.subplot(1,2,2).boxplot(bank_df['Income'])
plt.show()
```

**Ways to handle outliers**

- Remove the outliers data(not recommended)

- Fill with median values

- Cap the values with Q1 and Q3

```python
l1=[]
med=bank_df['Income'].median()
Q1=np.percentile(bank_df['Income'],25)
Q3=np.percentile(bank_df['Income'],75)
IQR=Q3-Q1
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR

for i in bank_df['Income']:
    if i<lb or i>ub:
        l1.append(med)
    else:
        l1.append(i)

bank_copy=bank_df.copy()
bank_copy['Income']=l1
```

```python
plt.subplot(1,2,1).boxplot(bank_copy['Income'])
plt.subplot(1,2,2).boxplot(bank_df['Income'])
plt.show()
```

## Standardization and Normalization

### To bring values under all columns in one scale

#### Standardization is measured by **ZScore**

#### Standardization value ranges from -3 to 3

<img src="https://fredclavel.org/wp-content/uploads/2019/03/z-score-population.png" jsaction="VQAsE" class="sFlh5c pT0Scc iPVvYb" style="max-width: 440px; height: 204px; margin: 0px; width: 353px;" alt="Basics: Standardization and the Z score | Fred Clavel, Ph.D." jsname="kn3ccd" aria-hidden="false">

```python
meann=bank_df['Income'].mean()
stdd=bank_df['Income'].std()
num= bank_df['Income']-meann
z=num/stdd
z
```

```python
bank_df['Income_Z']=z
```
